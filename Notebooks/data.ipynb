```python
file_path = 'Cyber_security.csv'  # Update this to the correct file path if necessary
df = pd.read_csv(file_path)

# Replace 'Asia-Pasific' with 'Asia-Pacific' in the 'region' column
df['Region'] = df['Region'].replace('Asia-Pasific', 'Asia-Pacific')

# Save the changes directly back to the same file
df.to_csv(file_path, index=False)

print("Replacement complete! The file has been updated.")
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('Cyber_security.csv')

# Renaming the column from 'NCSI' to 'NSCI'
df.rename(columns={'NCSI': 'NSCI'}, inplace=True)

# Save the updated DataFrame back to the same CSV file
df.to_csv('Cyber_security.csv', index=False)

# Display the updated DataFrame
print(df)
```
